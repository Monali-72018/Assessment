```python
# Healthcare Industry -

class ClinicAppointment:
    def __init__(self):
        self.appointments = {}

        self.time_slots = ["10am", "11am", "12pm", "2pm", "3pm"]

    # ---------------- BOOK APPOINTMENT ----------------
    def book_appointment(self):
        name = input("Enter patient name: ")
        age = int(input("Enter age: "))
        mobile = input("Enter mobile number: ")
        if len(mobile) != 10 or not mobile.isdigit():
            print("‚ùå Mobile number must be exactly 10 digits")
            return
        doctor = input("Enter preferred doctor: ")

        print("\nAvailable Time Slots:")
        for slot in self.time_slots:
            print(slot)

        slot = input("Choose time slot: ")

        if doctor not in self.appointments:
            self.appointments[doctor] = {}

        if slot not in self.appointments[doctor]:
            self.appointments[doctor][slot] = []

        if len(self.appointments[doctor][slot]) >= 3:
            print("  Slot full. Please choose another slot.")
            return

        patient = {
            "name": name,
            "age": age,
            "mobile": mobile,
            "doctor": doctor,
            "slot": slot
        }

        self.appointments[doctor][slot].append(patient)
        print("‚úÖ Appointment booked successfully!")

    # ---------------- VIEW APPOINTMENT ----------------
    def view_appointment(self):
        mobile = input("Enter mobile number to view appointment: ")

        for doctor in self.appointments:
            for slot in self.appointments[doctor]:
                for patient in self.appointments[doctor][slot]:
                    if patient["mobile"] == mobile:
                        print("\nüìÑ Appointment Details")
                        print("Name:", patient["name"])
                        print("Age:", patient["age"])
                        print("Doctor:", patient["doctor"])
                        print("Time Slot:", patient["slot"])
                        return

        print("‚ùå No appointment found.")

    # ---------------- CANCEL APPOINTMENT ----------------
    def cancel_appointment(self):
        mobile = input("Enter mobile number to cancel appointment: ")

        for doctor in self.appointments:
            for slot in self.appointments[doctor]:
                for patient in self.appointments[doctor][slot]:
                    if patient["mobile"] == mobile:
                        self.appointments[doctor][slot].remove(patient)
                        print("‚úÖ Appointment cancelled successfully!")
                        return

        print("‚ùå No appointment found to cancel.")
```

```python
clinic = ClinicAppointment()

while True:
    print("\n--- Clinic Appointment System ---")
    print("1. Book Appointment")
    print("2. View Appointment")
    print("3. Cancel Appointment")
    print("4. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
        clinic.book_appointment()
    elif choice == "2":
        clinic.view_appointment()
    elif choice == "3":
        clinic.cancel_appointment()
    elif choice == "4":
        print("Thank you! üëã")
        break
    else:
        print("Invalid choice!")
```

```python
# School Management System

class SchoolManagement:
    id_counter = 1

    def __init__(self):
        self.admissions = []

    # ---------------- NEW ADMISSION ----------------
    def new_admission(self):
        name = input("Enter student name: ")

        age = int(input("Enter age: "))
        if age < 5 or age > 18:
            print("‚ùå Age must be between 5 and 18")
            return

        student_class = int(input("Enter class (1‚Äì12): "))
        if student_class < 1 or student_class > 12:
            print("‚ùå Class must be between 1 and 12")
            return

        mobile = input("Enter guardian mobile number: ")
        if len(mobile) != 10 or not mobile.isdigit():
            print("‚ùå Mobile number must be exactly 10 digits")
            return

        student = {
            "id": SchoolManagement.id_counter,
            "name": name,
            "age": age,
            "class": student_class,
            "guardian_mobile": mobile
        }

        self.admissions.append(student)
        SchoolManagement.id_counter += 1

        print("‚úÖ New Admission Successful!")
        print("Student ID:", student["id"])

    # ---------------- VIEW STUDENT DETAILS ----------------
    def view_student_details(self):
        search_id = int(input("Enter student ID: "))

        for student in self.admissions:
            if student["id"] == search_id:
                print("\nüìÑ Student Details")
                print("ID:", student["id"])
                print("Name:", student["name"])
                print("Age:", student["age"])
                print("Class:", student["class"])
                print("Guardian Mobile:", student["guardian_mobile"])
                return

        print("‚ùå Student not found")

    # ---------------- UPDATE STUDENT INFO ----------------
    def update_student(self):
        search_id = int(input("Enter student ID to update: "))

        for student in self.admissions:
            if student["id"] == search_id:
                print("1. Update Class")
                print("2. Update Mobile Number")
                choice = int(input("Enter choice: "))

                if choice == 1:
                    new_class = int(input("Enter new class (1‚Äì12): "))
                    if 1 <= new_class <= 12:
                        student["class"] = new_class
                        print("‚úÖ Class updated successfully")
                    else:
                        print("‚ùå Invalid class")

                elif choice == 2:
                    new_mobile = input("Enter new mobile number: ")
                    if len(new_mobile) == 10 and new_mobile.isdigit():
                        student["guardian_mobile"] = new_mobile
                        print("‚úÖ Mobile number updated successfully")
                    else:
                        print("‚ùå Invalid mobile number")
                return

        print("‚ùå Student not found")

    # ---------------- REMOVE STUDENT RECORD ----------------
    def remove_student(self):
        search_id = int(input("Enter student ID to remove: "))

        for student in self.admissions:
            if student["id"] == search_id:
                self.admissions.remove(student)
                print("‚úÖ Student record removed successfully")
                return

        print("‚ùå Student not found")

    # ---------------- MENU SYSTEM ----------------
    def menu(self):
        while True:
            print("\n--- School Management System ---")
            print("1. New Admission")
            print("2. View Student Details")
            print("3. Update Student Info")
            print("4. Remove Student Record")
            print("5. Exit")

            choice = int(input("Enter your choice: "))

            if choice == 1:
                self.new_admission()
            elif choice == 2:
                self.view_student_details()
            elif choice == 3:
                self.update_student()
            elif choice == 4:
                self.remove_student()
            elif choice == 5:
                print("üëã Exiting system...")
                break
            else:
                print("‚ùå Invalid choice")
```

```python
school = SchoolManagement()
school.menu()
```

```python
# Transport Reservation System (Bus Ticketing)

class BusReservation:
    ticket_counter = 1001

    def __init__(self):
        self.routes = {
            "Mumbai to Pune": 500,
            "Delhi to Jaipur": 600,
            "Bangalore to Chennai": 700,
            "Ahmedabad to Surat": 400
        }

        self.tickets = []

        self.seats = {route: 0 for route in self.routes}

    # ---------------- SHOW AVAILABLE ROUTES ----------------
    def show_routes(self):
        print("\n Available Routes:")
        for route, price in self.routes.items():
            print(f"{route} - ‚Çπ{price}")

    # ---------------- BOOK TICKET ----------------
    def book_ticket(self):
        self.show_routes()

        route = input("\nEnter route exactly as shown: ")
        if route not in self.routes:
            print("‚ùå Invalid route")
            return

        if self.seats[route] >= 40:
            print("‚ùå No seats available for this route")
            return

        name = input("Enter passenger name: ")
        age = int(input("Enter age: "))

        mobile = input("Enter mobile number: ")
        if len(mobile) != 10 or not mobile.isdigit():
            print("‚ùå Invalid mobile number")
            return

        self.seats[route] += 1
        seat_number = self.seats[route]

        ticket = {
            "ticket_id": BusReservation.ticket_counter,
            "name": name,
            "age": age,
            "mobile": mobile,
            "route": route,
            "price": self.routes[route],
            "seat": seat_number
        }

        self.tickets.append(ticket)
        BusReservation.ticket_counter += 1

        print("\n‚úÖ Ticket Booked Successfully!")
        print("Ticket ID:", ticket["ticket_id"])
        print("Seat Number:", seat_number)

    # ---------------- VIEW TICKET ----------------
    def view_ticket(self):
        search_id = int(input("Enter ticket ID: "))

        for ticket in self.tickets:
            if ticket["ticket_id"] == search_id:
                print("\n Ticket Details")
                print("Ticket ID:", ticket["ticket_id"])
                print("Name:", ticket["name"])
                print("Age:", ticket["age"])
                print("Mobile:", ticket["mobile"])
                print("Route:", ticket["route"])
                print("Seat No:", ticket["seat"])
                print("Price: ‚Çπ", ticket["price"])
                return

        print("‚ùå Ticket not found")

    # ---------------- CANCEL TICKET ----------------
    def cancel_ticket(self):
        search_id = int(input("Enter ticket ID to cancel: "))

        for ticket in self.tickets:
            if ticket["ticket_id"] == search_id:
                self.tickets.remove(ticket)
                print("‚úÖ Ticket cancelled successfully")
                return

        print("‚ùå Ticket not found")

    # ---------------- MENU SYSTEM ----------------
    def menu(self):
        while True:
            print("\n--- Bus Reservation System ---")
            print("1. Show Available Routes")
            print("2. Book Ticket")
            print("3. View Ticket")
            print("4. Cancel Ticket")
            print("5. Exit")

            choice = int(input("Enter your choice: "))

            if choice == 1:
                self.show_routes()
            elif choice == 2:
                self.book_ticket()
            elif choice == 3:
                self.view_ticket()
            elif choice == 4:
                self.cancel_ticket()
            elif choice == 5:
                print("üëã Thank you for using Bus Reservation System")
                break
            else:
                print("‚ùå Invalid choice")
```

```python
bus = BusReservation()
bus.menu()
```
