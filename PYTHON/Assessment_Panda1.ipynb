```python
# 10. Are there outliers in GDP, Gold, or Oil prices?

cols = ['GDP Growth (%)','Gold Price (USD per Ounce)','Open Price']

Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

outliers = ((df[cols] < (Q1 - 1.5 * IQR)) |
                 (df[cols] > (Q3 + 1.5 * IQR))).sum()

outliers
```

```python
cols = ['GDP Growth (%)',
        'Gold Price (USD per Ounce)',
        'Open Price']

for col in cols:
    plt.figure()
    plt.boxplot(df[col].dropna())
    plt.title(f"Boxplot of {col}")
    plt.ylabel(col)
    plt.show()
```

```python
z_scores = np.abs((df[cols] - df[cols].mean()) / df[cols].std())

z_outliers = (z_scores > 3).sum()

plt.figure()
plt.bar(z_outliers.index, z_outliers.values)
plt.xlabel("Variables")
plt.ylabel("Number of Outliers (Z-score > 3)")
plt.title("Outliers Detection Using Z-Score Method")
plt.xticks(rotation=45)
plt.show()
```

```python
# Insightful Analysis Questions

# 1. What percentage of the dataset shows negative GDP growth?

total_count = len(df)
negative_count = (df['GDP Growth (%)'] < 0).sum()
percentage_negative = (negative_count / total_count) * 100

print("Percentage of Negative GDP Growth:",
      round(percentage_negative, 2), "%")
```

```python
# 2. Does high inflation correspond to higher interest rates?

correlation = df['Inflation Rate (%)'].corr(df['Interest Rate (%)'])

print("Correlation between Inflation and Interest Rate:",
      round(correlation, 3))


plt.figure()
plt.scatter(df['Inflation Rate (%)'], df['Interest Rate (%)'])
plt.xlabel("Inflation Rate (%)")
plt.ylabel("Interest Rate (%)")
plt.title("Inflation vs Interest Rate")
plt.show()
```

```python
# 3. Is there a relationship between unemployment and consumer spending?

subset = df[['Unemployment Rate (%)',
             'Consumer Spending (Billion USD)']].dropna()

correlation = subset['Unemployment Rate (%)'].corr(subset['Consumer Spending (Billion USD)'])

print("Correlation between Unemployment Rate (%) and Consumer Spending (Billion USD):",
      round(correlation, 3))


plt.figure()
plt.scatter(df['Unemployment Rate (%)'], df['Consumer Spending (Billion USD)'])
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Consumer Spending (Billion USD)")
plt.title("Unemployment Rate (%) vs Consumer Spending (Billion USD)")
plt.show()
```

```python
# 4. Do higher corporate profits align with higher consumer confidence?

correlation = df['Corporate Profits (Billion USD)'].corr(
              df['Consumer Confidence Index'])

print("Correlation between Corporate Profits and Consumer Confidence:",
      round(correlation, 3))


plt.figure()
plt.scatter(df['Corporate Profits (Billion USD)'],
            df['Consumer Confidence Index'])

plt.xlabel("Corporate Profits (Billion USD)")
plt.ylabel("Consumer Confidence Index")
plt.title("Corporate Profits vs Consumer Confidence")
plt.show()
```

```python
# 5. What’s the trend of crude oil prices over time?

df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')

plt.figure()
plt.plot(df['Date'], df['Close Price'])
plt.xlabel("Date")
plt.ylabel("Crude Oil Close Price")
plt.title("Trend of Crude Oil Prices Over Time")
plt.xticks(rotation=45)
plt.show()
```
