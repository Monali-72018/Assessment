```python
# Assessment 2: Retail Sales Analysis

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # basic charts
import seaborn as sns # advance charts like statatical
```

```python
df1 = pd.read_csv("/content/Retail Data.csv")
df1
```

```python
# 2. What is the shape (rows, columns) of the dataset?

print("Dataset Shape (rows, columns):", df1.shape)
```

```python
# 3. Are there any duplicate records?

duplicates = df1.duplicated().sum()
print("Number of duplicate records:", duplicates)
```

```python
# 4. Are there any missing or corrupted entries in Ship Date, Order Date, or numericcolumns?

print("Missing Values:\n")
print(df1[['Ship Date', 'Order Date']].isnull().sum())

numeric_cols = df1.select_dtypes(include='number').columns
print("\nMissing Values in Numeric Columns:\n")
print(df1[numeric_cols].isnull().sum())

df1['Ship Date'] = pd.to_datetime(df1['Ship Date'], errors='coerce')
df1['Order Date'] = pd.to_datetime(df1['Order Date'], errors='coerce')

print("\nCorrupted Ship Date entries:",
      df1['Ship Date'].isnull().sum())

print("Corrupted Order Date entries:",
      df1['Order Date'].isnull().sum())


for col in numeric_cols:
    converted = pd.to_numeric(df1[col], errors='coerce')
    if converted.isnull().sum() > df1[col].isnull().sum():
        print(f"Corrupted numeric values found in {col}")
```

```python
# 5. Convert Order Date and Ship Date to datetime.

df1['Order Date'] = pd.to_datetime(df1['Order Date'], errors='coerce')
df1['Ship Date'] = pd.to_datetime(df1['Ship Date'], errors='coerce')

print(df1[['Order Date', 'Ship Date']].dtypes)
```

```python
# 6. Check for future or inconsistent shipping dates.

invalid_shipping = df1[df1['Ship Date'] < df1['Order Date']]

print("Records where Ship Date is earlier than Order Date:",
      len(invalid_shipping))


future_ship = df1[df1['Ship Date'] > pd.Timestamp.today()]

print("Records with Future Ship Date:",
      len(future_ship))

future_order = df1[df1['Order Date'] > pd.Timestamp.today()]

print("Records with Future Order Date:",
      len(future_order))
```

```python
# 7.  Convert price columns to numeric (remove $ and commas).

df1['Cost Price'] = pd.to_numeric(df1['Cost Price'], errors='coerce')
df1['Retail Price'] = pd.to_numeric(df1['Retail Price'], errors='coerce')
df1['Profit Margin'] = pd.to_numeric(df1['Profit Margin'], errors='coerce')

print(df1[['Cost Price','Retail Price','Profit Margin']].dtypes)
```

```python
# 8. What are the unique values in Customer Type and Order Priority?

print("Unique Customer Types:")
print(df1['Customer Type'].unique())

print("Customer Type Counts:\n")
print(df1['Customer Type'].value_counts())

print("\nUnique Order Priorities:")
print(df1['Order Priority'].unique())

print("\nOrder Priority Counts:\n")
print(df1['Order Priority'].value_counts())
```

```python
# 9. What are the most common shipping modes?

most_common = df1['Ship Mode'].value_counts().idxmax()
count = df1['Ship Mode'].value_counts().max()

print("Most Common Shipping Mode:", most_common)
print("Count:", count)
```

```python
# 10. Which cities have the highest number of orders?

highest_number = df1['City'].value_counts().sort_index(ascending=False)
count = df1['City'].value_counts().max()

print("\nCity with Highest Orders:")
print(df1['City'].value_counts().head(5))
```

```python
# 11. What’s the range of order quantities and prices?

qty_range = df1['Order Quantity'].max() - df1['Order Quantity'].min()
price_range1 = df1['Cost Price'].max() - df1['Cost Price'].min()
price_range2 = df1['Retail Price'].max() - df1['Retail Price'].min()

print("Range of Order Quantity:", qty_range)
print("Range of Cost Price:", price_range1)
print("Range of Retail Price:", price_range2)
```

```python
# 12. Create a new column for shipping duration.

df1['shipping Duration'] = df1['Ship Date']-df1['Order Date']
df1
```

```python
# 13. Are there any orders with zero or negative total or quantity?

invalid_quantity = df1[df1['Order Quantity'] <= 0]

invalid_total = df1[df1['Cost Price'] <= 0]

print("Orders with zero or negative Quantity:", len(invalid_quantity))
print("Orders with zero or negative Total:", len(invalid_total))
```

```python
# 14. Are all discount percentages matching discount dollar amounts?

df1['Discount %'] = pd.to_numeric(df1['Discount %'], errors='coerce')

df1['Total'] = pd.to_numeric(df1['Total'], errors='coerce')

df1['Expected Discount'] = df1['Total'] * (df1['Discount %'] / 100)

df1['Match'] = np.isclose(
    df1['Expected Discount'],
    df1['Discount %'],
    atol=0.01
)

mismatched = df1[~df1['Match']]

print("Mismatched Records:", len(mismatched))
mismatched
```

```python
# 15.  Check for mismatches in total calculation’

df1['Retail Price'] = pd.to_numeric(df1['Retail Price'], errors='coerce')
df1['Order Quantity'] = pd.to_numeric(df1['Order Quantity'], errors='coerce')
df1['Total'] = pd.to_numeric(df1['Total'], errors='coerce')


df1['Expected Total'] = df1['Retail Price'] * df1['Order Quantity']

df1['Match'] = np.isclose(df1['Expected Total'],
                         df1['Total'],
                         atol=0.01)

mismatched = df1[~df1['Match']]

print("Total Records:", len(df1))
print("Mismatched Total Records:", len(mismatched))

print("\nSample Mismatches:")
mismatched[['Retail Price',
                  'Order Quantity',
                  'Total',
                  'Expected Total']].head()
```

```python
# 16.  Identify top 5 products by order quantity.

df1.columns = df1.columns.str.strip()

top_products = (
    df1.groupby('Product Name')['Order Quantity']
      .sum()
      .sort_values(ascending=False)
      .head(5)
)

print("Top 5 Products by Order Quantity:\n")
top_products
```

```python
# 17. Which Account Manager handled the most revenue?

df1['Total'] = pd.to_numeric(df1['Total'], errors='coerce')

revenue_by_manager = (
    df1.groupby('Account Manager')['Total']
       .sum()
       .sort_values(ascending=False)
)

top_manager = revenue_by_manager.idxmax()
top_revenue = revenue_by_manager.max()

print("Account Manager with Highest Revenue:", top_manager)
print("Total Revenue:", top_revenue)
```

```python
# 18.  What is the average shipping cost by mode?

df1['Shipping Cost'] = pd.to_numeric(df1['Shipping Cost'], errors='coerce')

avg_shipping = (
    df1.groupby('Ship Mode')['Shipping Cost']
       .mean()
       .sort_values(ascending=False)
)

print("Average Shipping Cost by Mode:\n")
print(avg_shipping)
```

```python
# Insightful Analysis Questions

# 1. What is the total revenue generated across all orders?

df1['Total'] = pd.to_numeric(df1['Total'], errors='coerce')

total_revenue = df1['Total'].sum()

print("Total Revenue Generated:", total_revenue)
```

```python
# 2. Which customer type generates more revenue?

df1['Total'] = pd.to_numeric(df1['Total'], errors='coerce')

df_clean = df1.dropna(subset=['Total'])

revenue_by_customer = (
    df_clean.groupby('Customer Type')['Total']
            .sum()
            .sort_values(ascending=False)
)

if not revenue_by_customer.empty:
    print(revenue_by_customer)
    print("\nTop Customer Type:", revenue_by_customer.idxmax())
else:
    print("No valid revenue data found.")
```

```python
# 3. How does order priority affect revenue?

df1['Total'] = pd.to_numeric(df1['Total'], errors='coerce')

priority_revenue = (
    df1.groupby('Order Priority')['Total']
       .sum()
       .sort_values(ascending=False)
)

print("Revenue by Order Priority:\n")
priority_revenue
```

```python
# 4. What is the average profit margin by product category?

df1['Profit Margin'] = pd.to_numeric(df1['Profit Margin'], errors='coerce')
df1['Total'] = pd.to_numeric(df1['Total'], errors='coerce')

df1 = df1[df1['Total'] != 0]

df1['Profit Margin %'] = (df1['Profit Margin'] / df1['Total']) * 100

avg_margin = (
    df1.groupby('Product Category')['Profit Margin %']
       .mean()
       .sort_values(ascending=False)
)

print("Average Profit Margin by Category:\n")
avg_margin
```

```python
# 5. What is the most profitable product overall?

df1['Profit Margin'] = pd.to_numeric(df1['Profit Margin'], errors='coerce')

profit_by_product = (
    df1.groupby('Product Name')['Profit Margin']
       .sum()
       .sort_values(ascending=False)
)

most_profitable_product = profit_by_product.idxmax()
max_profit = profit_by_product.max()

print("Most Profitable Product:", most_profitable_product)
print("Total Profit:", max_profit)
```

```python
# 6. How many days does it usually take to ship an order?

df1['Order Date'] = pd.to_datetime(df1['Order Date'], errors='coerce')
df1['Ship Date'] = pd.to_datetime(df1['Ship Date'], errors='coerce')

df1['Shipping Days'] = (df1['Ship Date'] - df1['Order Date']).dt.days

avg_shipping_days = df1['Shipping Days'].mean()

print("Average Shipping Days:", avg_shipping_days)
```

```python
# 7. Do longer shipping times impact profit margins?

df1['Profit Margin'] = pd.to_numeric(df1['Profit Margin'], errors='coerce')
df1['Total'] = pd.to_numeric(df1['Total'], errors='coerce')

df1 = df1[df1['Total'] != 0]
df1['Profit Margin %'] = (df1['Profit Margin'] / df1['Total']) * 100

correlation = df1['Shipping Days'].corr(df1['Profit Margin %'])

print("Correlation between Shipping Days & Profit Margin:", correlation)
```

```python
# 8. Which city brings in the highest revenue?

df1['Total'] = pd.to_numeric(df1['Total'], errors='coerce')

city_revenue = (
    df1.groupby('City')['Total']
       .sum()
       .sort_values(ascending=False)
)

print("Top Revenue City:", city_revenue.idxmax())
```

```python
# 9. Which account manager generated the most revenue?

manager_revenue = (
    df1.groupby('Account Manager')['Total']
       .sum()
       .sort_values(ascending=False)
)

print("Top Manager:", manager_revenue.idxmax())
```

```python
# 10.  Which shipping mode is most cost-effective (lowest avg. shipping)?

df1['Shipping Cost'] = pd.to_numeric(df1['Shipping Cost'], errors='coerce')

shipping_cost = (
    df1.groupby('Ship Mode')['Shipping Cost']
       .mean()
       .sort_values()
)

print("Most Cost-Effective Shipping Mode:", shipping_cost.idxmin())
```

```python
# 11.  Do higher discounts reduce profits?

df1['Discount %'] = pd.to_numeric(df1['Discount %'], errors='coerce')

discount_profit_corr = df1['Discount %'].corr(df1['Profit Margin'])

print("Correlation between Discount & Profit:", discount_profit_corr)
```

```python
# 12. Which state has the highest number of orders?

state_orders = df1['State'].value_counts()

print("State with Highest Orders:", state_orders.idxmax())
```

```python
# 13. What is the average discount % across all orders?

avg_discount = df1['Discount %'].mean()

print("Average Discount %:", avg_discount)
```

```python
# 14. What is the average total spend per order?

avg_spend = df1['Total'].mean()

print("Average Total Spend per Order:", avg_spend)
```

```python
# 15. . Are certain containers (e.g., Small Box, Wrap Bag) more profitable?

container_profit = (
    df1.groupby('Product Container')['Profit Margin']
       .sum()
       .sort_values(ascending=False)
)

print("Most Profitable Container Type:", container_profit.idxmax())
```
